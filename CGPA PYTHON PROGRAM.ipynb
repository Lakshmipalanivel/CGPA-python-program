```python
#cgpa calculator
def cgpa_display():
    total1=0
    gtotal=0
    grades={"O":"10","A+":"9","A":"8","B+":"7","B":"6","U":"0"}
    name=input("\n Enter the student name:")
    dept=input("\n Enter department of the student:")
    year=input("\n Enter year of the student:")
    sem=input("\n Enter semester of the student:")
    n=int(input("\n Enter the no.of.subjects:"))
    print("\n Enter the grades in (O-U)U-arrear:")
    marks=[]
    for entry in range(n):
        cc=input("Enter the coursecode:")
        ma=input("Enter the grades in (O-U):")
        g=float(input("Enter the gradepoints:"))
        if ma in grades:
            m=float(grades[ma])
            ma=ma.upper()
            entry=(cc,g,m,ma)
            marks.append(entry)
            if ma=="U":
                mn="0"
            else:
                mn="1"
    if mn=="0":
         print("\n\tNAME:",name,"\t\tDEPARTMENT:",dept)
         print("\n\tSEMESTER:",sem,"\t\tYEAR:",year)
         print("\ncoursecode grade")
         for entry in marks:
             cc,g,m,ma=entry
             print("\n\t",cc, "\n\t",ma)
         print("\n\nCGPA=ARREAR")
    else:
        print("\n\tNAME:",name,"\t\tDEPARTMENT:",dept)
        print("\n\tSEMESTER:",sem,"\t\tYEAR:",year)
        print("\ncoursecode grade")
        for entry in marks:
            cc,g,m,ma=entry
            print("\n\t",cc, "\n\t",ma)
            total1=total1+m*g
            gtotal=gtotal+g
        cgpa=total1/gtotal
        print("\n CGPA=",cgpa)
cgpa_display()
```
